# Exploratory Data Analysis

This notebook examines employee turnover and highlights potential drivers using HR data.


## Objective

Explore employee turnover and identify factors that may impact retention.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Loading and Initial Overview

The dataset is loaded to begin the analysis.


```python
df = pd.read_csv("../data/raw/hr_data_raw.csv")
df.head()
```

## Dataset Structure

This section summarizes the size, structure, and data types.


```python
df.shape
df.info()
```

## Attrition Overview

This section analyzes overall employee turnover.


```python
df["Termd"].value_counts()
df["Termd"].value_counts(normalize=True)
```

```python
sns.countplot(data=df, x="Termd")
plt.title("Employee Attrition")
plt.show()
```

## Attrition by department

Certain departments show higher employee turnover, suggesting potential organizational or workload-related issues.

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Department", hue="Termd")
plt.title("Attrition by Department")
plt.xticks(rotation=45)
plt.show()
```

## Attrition Drivers


### Attrition and Salary

Salary distributions are compared between active and former employees.


```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Termd", y="Salary")
plt.title("Salary distribution by Attrition")
plt.show()
```

### Attrition and performance

Performance ratings are analyzed in relation to attrition.


```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="PerformanceScore", hue="Termd")
plt.title("Attrition by Performance Score")
plt.xticks(rotation=45)
plt.show()
```

#### Performance Score Distribution 

Employees with lower performance scores are more frequently observed among those who left the company.

## Key Insights

- Employee turnover is relatively high.
- Production and Sales show the highest attrition.
- Lower salaries are more common among employees who left.
- Lower performance ratings are associated with higher attrition.
